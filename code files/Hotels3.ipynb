```python
from bs4 import BeautifulSoup
import requests
import pandas
from openpyxl import load_workbook
```

```python
wb = load_workbook(filename = 'links.xlsx')
print(wb)
```

```python
sheet_ranges = wb['sheet3']
print(sheet_ranges)
```

```python
names = []
for row in sheet_ranges.iter_rows(): 
    for cell in row:
        cell.value = cell.value.replace(' ','_')
        names.append(cell.value)
print(names)
```

```python
urls = []
for name in names:
    url = 'https://www.google.com/travel/hotels/'+name+'?'
    urls.append(url)

for index,url in enumerate(urls):
    print(index+1,url)
```

```python
def getHotels(url):
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text,'lxml')
    tab = soup.find('div',class_='l5cSPd')
    container = tab.find_all('c-wiz',class_='f1dFQe')
    name = ''
    for div in container[:3]:
        hotel = div.find('div',class_='QT7m7')
        if hotel:
            name+=hotel.text+' | '
        else:
            continue
    if len(name) > 1:
        return name
    else:
        return 'none'
```

```python
for index,url in enumerate(urls):
    print(getHotels(url))
```
