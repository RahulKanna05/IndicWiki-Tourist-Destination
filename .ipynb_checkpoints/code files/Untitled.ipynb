```python
def getType(url):
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text,'lxml')
    table = soup.find('table',class_="infobox")
    if table:
        tr_list = table.find_all('tr')
        for tr in tr_list:
            if tr.th and tr.th.text == 'Type':
                return tr.td.text
    
```

```python
for index,url in enumerate(us_list):
    print(getType(url))
```

```python
def getVisitors(url):
    html_text = requests.get(url).text
    soup = BeautifulSoup(html_text,'lxml')
    table = soup.find('table',class_="infobox")
    if table:
        tr_list = table.find_all('tr')
        for tr in tr_list:
            if tr.th and 'Visitors' in tr.th.text:
                return tr.td.text
```

```python
for index,url in enumerate(us_list):
    print(getVisitors(url))
```
